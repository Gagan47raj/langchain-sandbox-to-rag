```python
from langchain_community.document_loaders import TextLoader
```

```python
# 1. Initialize Loader
# Note: Always specify encoding to prevent Windows/Linux UTF-8 crashes
loader = TextLoader("dataset/server_logs.txt", encoding="utf-8")
```

```python
# 2. Load the file
docs = loader.load()
```

```python
# 3. Inspecting Document and Metadata
print(f"Total documents loaded: {len(docs)}")
```

```python
first_doc = docs[0]
print(f"Metadata: {first_doc.metadata}")
```

```python
print(f"Content snippet: {first_doc.page_content[:100]}...")
```

```python
from langchain_community.document_loaders import CSVLoader
```

```python
# 1. Initialize Loader
loader = CSVLoader(file_path="dataset/patient_records.csv")
```

```python
# 2. Load Data
docs = loader.load()
```

```python
# 3. Inspecting the output
print(f"Total rows (documents) loaded: {len(docs)}")
```

```python
print(f"First Row Metadata: {docs[0].metadata}")
```

```python
print(f"First Row Content:\n{docs[0].page_content}")
```

```python
from langchain_community.document_loaders import PyPDFLoader
```

```python
# 1. Initialize
loader = PyPDFLoader("dataset/medical_imaging_handbook.pdf")
```

```python
# 2. Load (Extracts page by page)
docs = loader.load()
```

```python
# 3. Inspect
print(f"Total pages loaded: {len(docs)}")
```

```python
page_5 = docs[4] # 0-indexed
print(f"Page 5 Metadata: {page_5.metadata}")
```

```python
print(f"Page 5 Content snippet: {page_5.page_content[:200]}")
```
