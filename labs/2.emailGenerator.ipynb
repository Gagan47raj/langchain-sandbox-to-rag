```python
from langchain_ollama import ChatOllama
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
# 1. Initialize Ollama Mistral Model
llm = ChatOllama(
    model="mistral",
    temperature=0.7
)
```

```python
# 2. Create Prompt Template
template = """
You are a professional email generator.

Write a formal and professional email from {sender} regarding {purpose}.
Include a suitable subject line, greeting, email body, and closing.
"""
```

```python
prompt_template = PromptTemplate.from_template(template)
```

```python
# 3. Output Parser
parser = StrOutputParser()
```

```python
# 4. Build Chain using LCEL
chain = prompt_template | llm | parser
```

```python
# 5. Execute Chain
if __name__ == "__main__":
    purpose = "Asking for a 2-day leave due to a family trip."

    print("Generating email...")

    result = chain.invoke({
        "sender": "Gagan",
        "purpose": purpose
    })

    print("\nFinal Result:")
    print(result)
```
