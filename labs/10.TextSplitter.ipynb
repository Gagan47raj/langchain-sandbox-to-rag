```python
from langchain_text_splitters import RecursiveCharacterTextSplitter, CharacterTextSplitter
from langchain_core.documents import Document
```

```python
# 1. Dummy Large Text (Simulating a long document)
long_text = """
LangChain is a framework designed to simplify the creation of applications using LLMs.
It allows developers to connect language models to other sources of data.

Chunking is a very important part of this process. If you don't chunk, you will hit context limits.
Always ensure you have some overlap so sentences are not completely lost in translation.
"""
```

```python
# 2. Initialize the Recursive Splitter
# In production, chunk_size is usually 500-1000. We use small numbers here for demonstration.
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=100,         # Maximum characters per chunk
    chunk_overlap=20,       # Number of characters overlapping between chunks
    separators=["\n\n", "\n", " ", ""] # The fallback order
)
```

```python
# 3. Create a Document Object (Simulating output from a Document Loader)
doc = Document(page_content=long_text, metadata={"source": "dataset/langchain_guide.txt"})
```

```python
# 4. Perform the Split
print("Splitting Document...\n")
chunks = text_splitter.split_documents([doc])
```

```python
print(f"Total Chunks Created: {len(chunks)}\n")
```

```python
for i, chunk in enumerate(chunks):
    print(f"--- Chunk {i+1} ---")
    print(f"Length: {len(chunk.page_content)} characters")
    print(f"Content: {chunk.page_content}")
    print(f"Metadata: {chunk.metadata}\n")
```

```python
from langchain_community.document_loaders import TextLoader, PyPDFLoader
```

```python
# Load file
loader1 = TextLoader("dataset/server_logs.txt")
loader2 = PyPDFLoader("dataset/medical_imaging_handbook.pdf")
```

```python
documents1 = loader1.load()
documents2 = loader2.load()
```

```python
print("Documents Loaded:", len(documents1))
print("Documents Loaded:", len(documents2))
```

```python
# Splitters
splitter1 = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

splitter2 = CharacterTextSplitter(
    separator=".",
    chunk_size=50,
    chunk_overlap=10
)
```

```python
# Create chunks
chunks1 = splitter1.split_documents(documents1)
chunks2 = splitter2.split_documents(documents2)
```

```python
print("\nServer Log Chunks:", len(chunks1))
print("Medical PDF Chunks:", len(chunks2))
```

```python
# Preview first few server log chunks
for i, chunk in enumerate(chunks1[:5]):
    print(f"\n--- Chunk {i+1} ---")
    print("Source:", chunk.metadata.get("source"))
    print(chunk.page_content[:200])
```

```python
# Preview first few server log chunks
for i, chunk in enumerate(chunks2[:5]):
    print(f"\n--- Chunk {i+1} ---")
    print("Source:", chunk.metadata.get("source"))
    print(chunk.page_content[:200])
```

```python
text = """
This is a very long sentence that intentionally exceeds fifty characters in length so we can test the splitter.
Short sentence.
"""

splitter = CharacterTextSplitter(
    separator=".",
    chunk_size=50,
    chunk_overlap=0
)

chunks = splitter.split_text(text)

for chunk in chunks:
    print(chunk)
    print("Length:", len(chunk))
```
