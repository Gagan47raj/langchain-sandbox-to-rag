```python
from langchain_ollama import ChatOllama
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
# 1. Initialize Ollama Mistral Model
llm = ChatOllama(
    model="mistral",
    temperature=0.7
)
```

```python
# 2. Create Prompt Template
template = """
You are a branding expert.
Suggest a catchy company name for a startup that makes {company_product}.
"""
```

```python
prompt_template = PromptTemplate.from_template(template)
```

```python
# 3. Output Parser
parser = StrOutputParser()
```

```python
# 4. Build Chain using LCEL
chain = prompt_template | llm | parser
```

```python
# 5. Execute Chain
if __name__ == "__main__":
    product_type = "AI-powered medical devices"

    print("Generating name...")

    result = chain.invoke({
        "company_product": product_type
    })

    print("\nFinal Result:")
    print(result)
```
