```python
from collections import Counter

extensions = Counter()

domains = Counter()

for doc in documents:

    extensions[doc.metadata["extension"]] += 1
    domains[doc.metadata["domain"]] += 1

print("Extension Distribution")

print(extensions)

print()

print("Domain Distribution")

print(domains)
```
