```python
from langchain_ollama import ChatOllama
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python

# 1. Initialize components
llm = ChatOllama(
    model="mistral",
    temperature=0.7
)
```

```python

parser = StrOutputParser()
```

```python
# 2. Create Prompt Template
template = "Write a catchy 2-line intro for a Gobus Billa YouTube video about {topic}."

prompt = PromptTemplate.from_template(template)
```

```python
# 3. Build the RunnableSequence using LCEL
modern_chain = prompt | llm | parser
```

```python
# 4. Execute the chain
print("\n--- LCEL / RunnableSequence Output ---")
```

```python
result = modern_chain.invoke({
    "topic": "Linked Lists vs Arrays"
})
```

```python
print(result)
```
