```python
from langchain_ollama import ChatOllama
from langchain_core.prompts import PromptTemplate
from langchain_core.runnables import RunnableLambda
```

```python
# 1. Initialize LLM
llm = ChatOllama(
    model="mistral",
    temperature=0.7
)
```

```python
# 2. Prompt Template
recipe_prompt = PromptTemplate.from_template(
    "Generate a recipe for {dish}."
)
```

```python
# 3. Custom Parser using Python string method
uppercase_parser = RunnableLambda(
    lambda msg: msg.content.upper()
)
```

```python
# Build LCEL Chain
recipe_chain = recipe_prompt | llm | uppercase_parser
```

```python
# Execute
result = recipe_chain.invoke({
    "dish": "Paneer Butter Masala"
})
```

```python
print(result)
```

```python
from langchain_core.output_parsers import StrOutputParser
```

```python
uppercase_parser = RunnableLambda(
    lambda text: text.upper()
)
```

```python
recipe_chain = (
    recipe_prompt
    | llm
    | StrOutputParser()
    | uppercase_parser
)
```

```python
result = recipe_chain.invoke({
    "dish": "Paneer Butter Masala"
})

```

```python
print(result)
```
