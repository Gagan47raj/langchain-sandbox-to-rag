```python
from langchain_core.documents import Document

from langchain_ollama import OllamaEmbeddings

from langchain_community.vectorstores import FAISS
from langchain_chroma import Chroma
```

```python
texts = [
    "A happy dog is playing in the park.",

    "A joyful puppy is running around the garden.",

    "I bought a brand new Apple MacBook Pro.",

    "The new iPhone 15 features an amazing camera.",

    "Stock market crashed heavily today due to inflation."
]

docs = [
    Document(page_content=text)
    for text in texts
]
```

```python
embeddings = OllamaEmbeddings(
    model="nomic-embed-text"
)
```

```python
print("=" * 80)
print("FAISS VECTOR STORE")
print("=" * 80)

faiss_db = FAISS.from_documents(
    docs,
    embeddings
)
```

```python
query = "Tell me about a cheerful canine."

results = faiss_db.similarity_search_with_score(
    query,
    k=2
)

print("Query:", query)
print()

for doc, score in results:
    print("Document:")
    print(doc.page_content)
    print(f"Distance Score: {score:.4f}")
    print("-" * 50)
```

```python
print("=" * 80)
print("CHROMA VECTOR STORE")
print("=" * 80)

chroma_db = Chroma.from_documents(
    documents=docs,
    embedding=embeddings,
    persist_directory="chroma_storage"
)
```

```python
query = "Are there any updates on tech gadgets?"

results = chroma_db.similarity_search_with_score(
    query,
    k=2
)

print("Query:", query)
print()

for doc, score in results:
    print("Document:")
    print(doc.page_content)
    print(f"Distance Score: {score:.4f}")
    print("-" * 50)
```
