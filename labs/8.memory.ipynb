```python
from langchain_ollama import ChatOllama

from langchain_core.prompts import (
    ChatPromptTemplate,
    MessagesPlaceholder
)

from langchain_core.chat_history import (
    InMemoryChatMessageHistory
)

from langchain_core.runnables.history import (
    RunnableWithMessageHistory
)
```

```python
# 1. Initialize LLM
llm = ChatOllama(
    model="mistral",
    temperature=0.7
)
```

```python
# 2. Prompt Template with Memory Placeholder
prompt = ChatPromptTemplate.from_messages([
    (
        "system",
        "You are an expert AI software engineering mentor."
    ),

    MessagesPlaceholder(
        variable_name="chat_history"
    ),

    (
        "human",
        "{input}"
    )
])
```

```python

# 3. Standard LCEL Chain
chain = prompt | llm
```

```python
# 4. Session Store (acts like memory database)
session_store = {}


def get_session_history(session_id: str):
    if session_id not in session_store:
        session_store[session_id] = (
            InMemoryChatMessageHistory()
        )

    return session_store[session_id]
```

```python
# 5. Add Message History
memory_chain = RunnableWithMessageHistory(
    chain,
    get_session_history,
    input_messages_key="input",
    history_messages_key="chat_history"
)
```

```python
# -----------------------------
# Execution
# -----------------------------
if __name__ == "__main__":

    config = {
        "configurable": {
            "session_id": "user_session_101"
        }
    }

    # Turn 1
    response_1 = memory_chain.invoke(
        {
            "input":
            "Hi, my favorite sorting algorithm is Merge Sort."
        },
        config=config
    )

    print("AI:", response_1.content)
    print()

    # Turn 2
    response_2 = memory_chain.invoke(
        {
            "input":
            "What is my favorite algorithm?"
        },
        config=config
    )

    print("AI:", response_2.content)
```

```python
from langchain_core.output_parsers import StrOutputParser
```

```python
# Add parser
parser = StrOutputParser()

# Updated chain
chain = prompt | llm | parser
```

```python
memory_chain = RunnableWithMessageHistory(
    chain,
    get_session_history,
    input_messages_key="input",
    history_messages_key="chat_history"
)
```

```python
# Session 101
config_101 = {
    "configurable": {
        "session_id": "user_session_101"
    }
}
```

```python
# Session 102
config_102 = {
    "configurable": {
        "session_id": "user_session_102"
    }
}
```

```python
# Store information in Session 101
memory_chain.invoke(
    {"input": "My favorite sorting algorithm is Merge Sort."},
    config=config_101
)
```

```python
# Ask in Session 101
response = memory_chain.invoke(
    {"input": "What is my favorite algorithm?"},
    config=config_101
)
```
