```python
from langchain_core.prompts import PromptTemplate
```

```python
# ==========================================
# Concept 7: Single Variable PromptTemplate
# ==========================================
print("--- Single Variable Example ---")
# 1. Define template with one placeholder {topic}
single_template = "Explain the core concept of {topic} in 2 simple sentences."

# 2. Create PromptTemplate object
prompt_1 = PromptTemplate.from_template(single_template)

# 3. Format the prompt dynamically
formatted_1 = prompt_1.format(topic="Array Data Structures")
print(f"Generated Prompt:\n{formatted_1}\n")
```

```python
# ==========================================
# Concept 8 & 9: Multiple Variable & Dynamic Prompts
# ==========================================
print("--- Multiple Variable Example ---")
# Let's say you are creating dynamic content based on user inputs
multi_template = """
You are an expert technical instructor. 
Write a short YouTube intro script for a video about {topic}.
The target audience is {audience}.
Keep the tone {tone}.
"""

# We can explicitly define variables for strict validation
prompt_2 = PromptTemplate(
    template=multi_template,
    input_variables=["topic", "audience", "tone"]
)

# Formatting dynamically
formatted_2 = prompt_2.format(
    topic="Depth First Search (DFS)",
    audience="first-year engineering students",
    tone="encouraging and energetic"
)
print(f"Generated Prompt:\n{formatted_2}")
```

```python
template = "Show me the setup commands for {framework} in {language}."
prompt_template = PromptTemplate.from_template(template)
result = prompt_template.invoke({
    "framework": "Django",
    "language": "Python"
})

print(result.text)
```

```python


template = """
You are a setup assistant.

Provide setup commands for {framework} in {language}.

Return the answer in this JSON format:
{{
  "framework": "name of framework",
  "language": "programming language",
  "commands": [
    "command1",
    "command2"
  ]
}}
"""

prompt_template = PromptTemplate.from_template(template)

prompt = prompt_template.invoke({
    "framework": "React",
    "language": "JavaScript" 
})

print(prompt.text)
```

```python
# Create PromptTemplate with strict input variables
template = """
You are a medical diagnostic assistant.

Create a structured clinical intake summary using the following details:

Patient Age: {patient_age}
Symptoms: {symptoms}
Duration of Symptoms: {duration}

Format the output professionally with sections:
- Patient Information
- Presenting Complaints
- Symptom Duration
- Clinical Summary
"""
```

```python
prompt_template = PromptTemplate(
    input_variables=["patient_age", "symptoms", "duration"],
    template=template
)
```

```python
# Pass variables using .format()
final_prompt = prompt_template.format(
    patient_age="35 years",
    symptoms="Fever, headache, and sore throat",
    duration="3 days"
)

# Print final prompt
print(final_prompt)
```
